# Data information:

#Num: Represents a numerical identifier for each row.
#Networth_Next_Year: Represents the net worth for the next year. This appears to be the target variable for prediction.
#Total_assets: Total assets of the company.
#Net_worth: Net worth of the company.
#Total_income: Total income earned by the company.
#Change_in_stock: Change in stock value.
#Total_expenses: Total expenses incurred by the company.
#Profit_after_tax: Profit earned by the company after tax deductions.
#PBDITA: Represents Profit Before Depreciation, Interest, Taxes, and Amortization.
#PBT: Represents Profit Before Tax.
#Finished_goods_turnover: Turnover of finished goods.
#WIP_turnover: Work-in-progress turnover.
#Raw_material_turnover: Turnover of raw materials.
#Shares_outstanding: Total number of outstanding shares.
#Equity_face_value: Face value of equity.
#EPS: Earnings Per Share.
#Adjusted_EPS: Adjusted Earnings Per Share.
#Total_liabilities: Total liabilities of the company.
#PE_on_BSE: Price to Earnings ratio on the Bombay Stock Exchange.
#Target: Binary column indicating the target class or category. It seems to represent whether a company is a target for some specific analysis or prediction task.

```python
# Arrange columns in better way
df.columns=df.columns.str.replace(' ','_').str.replace('%','_').str.replace('/','_').str.replace('(','_').str.replace(')','_')
```

```python
# Create Classification Target Variable
df['Target']=np.where((df['Networth_Next_Year']>0),1,0)
```

```python
df['Target'].value_counts()
```

```python
df.head()
```

```python
Target=df['Target']
f_col=df.drop('Target',axis=1)
```

```python
# checking null value
sorted=f_col.isnull().sum()*100/len(f_col)
print(sorted.sort_values(ascending=False))
```

```python
sns.heatmap(f_col.isnull(),cmap='coolwarm')
```

```python
# Checking Outliers
for i in f_col:
  sns.boxplot(f_col[i].dropna())
  plt.show()
```

```python
Q1=f_col.quantile(0.25)
Q3=f_col.quantile(0.75)
IQR=Q3-Q1
Upper_Side=Q3+1.5*IQR
Lower_Side=Q1-1.5*IQR
print('Upper_Side\n\n',Upper_Side,'Lower_Side \n\n',Lower_Side)
```

```python
((f_col<Lower_Side) | (f_col>Upper_Side)).sum()
```

```python
f_col[((f_col<Lower_Side) | (f_col>Upper_Side))]=np.nan
```

```python
sns.heatmap(f_col.isnull(),cmap='coolwarm')
# Null Value increse because we convert outlier to nan value
```

```python
f_col=pd.concat([Target,f_col],axis=1)
```

```python
f_col.shape
```

```python
""" Deposits__accepted_by_commercial_banks_        100.000000
 PE_on_BSE                                       61.959898
 Investments                                     40.525275
 Other_income                                    36.571590
 Contingent_liabilities                          33.549845
 Deferred_tax_liability                          32.194295
 Income_from_financial_services                  26.404970"""
```

```python
# We delet 25% null Value
f_col.drop(['Networth_Next_Year','Num','Deposits__accepted_by_commercial_banks_','PE_on_BSE','Investments','Other_income','Contingent_liabilities','Deferred_tax_liability','Income_from_financial_services'],inplace=True,axis=1)
```

```python
f_col.shape
```

```python
#Fill Null Value
from sklearn.impute import KNNImputer
Imputer=KNNImputer()
f_col_Imputer=pd.DataFrame(Imputer.fit_transform(f_col),columns=f_col.columns)
```
